# Conditional cross-sectional rank: Oracle vs HITS

Anchored WFO on 1T, 2021–2025. Oracle and HITS long/short strategies are each tested with and without a conditional cross-sectional ranker. The ranker is an entry selector only; Oracle sell/cover and HITS authority exits remain unchanged. Singleton cross-sections are excluded from rank training.

```python
import os
os.environ['CROSS_RANK_OUT'] = '../artifacts/cross_rank_oracle_hits_anchored_wfo_notebook'
os.environ['CROSS_RANK_TIERS'] = '1T'
os.environ['CROSS_RANK_FIRST_YEAR'] = '2021'
os.environ['CROSS_RANK_LAST_YEAR'] = '2025'
os.environ['CROSS_RANK_TOP_K'] = '10'
os.environ['CROSS_RANK_RF_ESTIMATORS'] = '40'
%run ../scripts/run_cross_rank_oracle_hits_anchored_wfo.py
```

```python
import pandas as pd
results = pd.read_parquet('../artifacts/cross_rank_oracle_hits_anchored_wfo_notebook/1t_results.parquet')
summary = (results.groupby(['model', 'variant', 'ranker'])
           .agg(avg_annual_return=('total_return', 'mean'), median_return=('total_return', 'median'),
                min_return=('total_return', 'min'), max_return=('total_return', 'max'),
                avg_sharpe=('sharpe', 'mean'))
           .reset_index().round(4))
summary
```
