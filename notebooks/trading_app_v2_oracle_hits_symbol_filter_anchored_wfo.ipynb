# Oracle/HITS anchored WFO with per-symbol profitability filtering

For forward year `Y`, RF models are trained on all data through `Y-1`. Each symbol is backtested independently on `Y-1` with `backtesting.py`; only symbols with a profitable prior-year strategy are traded in `Y`. Oracle and sparse-HITS labels are evaluated separately for long-only and short-only strategies.

The default run covers 1T, 100B, and 10B for 2021–2025. Environment variables can restrict tiers, years, or feature families for a smoke test.

```python
import os
os.environ.setdefault('SYMBOL_FILTER_TIERS', '1T,100B,10B')
os.environ.setdefault('SYMBOL_FILTER_FIRST_YEAR', '2021')
os.environ.setdefault('SYMBOL_FILTER_LAST_YEAR', '2025')
os.environ.setdefault('SYMBOL_FILTER_RF_ESTIMATORS', '40')

%run ../scripts/run_symbol_filter_anchored_wfo.py
```
