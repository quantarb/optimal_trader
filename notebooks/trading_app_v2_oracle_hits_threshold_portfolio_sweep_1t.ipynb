# 1T Oracle vs HITS threshold and portfolio-cap sweep

Anchored WFO trains through the prior year and evaluates 2021–2025. The same trained predictions are backtested across entry/exit thresholds and shared-book top-k portfolio caps. This is long-only and covers Oracle and the two-model sparse-HITS strategy.

```python
import os
os.environ['WFO_OUT'] = '../artifacts/oracle_hits_threshold_portfolio_sweep_1t'
os.environ['WFO_TIERS'] = '1T'
os.environ['WFO_FIRST_TEST_YEAR'] = '2021'
os.environ['WFO_LAST_TEST_YEAR'] = '2025'
os.environ['WFO_VARIANTS'] = 'long'
os.environ['WFO_ORACLE_THRESHOLDS'] = '0.5,0.6,0.7'
os.environ['WFO_HITS_THRESHOLDS'] = '0.8,0.85,0.9'
os.environ['WFO_TOP_KS'] = '5,10,20,40'
%run ../scripts/run_oracle_hits_anchored_wfo.py
```

```python
import pandas as pd
results = pd.read_parquet('../artifacts/oracle_hits_threshold_portfolio_sweep_1t/1t_results.parquet')
summary = (results.groupby(['model', 'entry_threshold', 'portfolio_top_k', 'year'])
           .agg(mean_return=('total_return', 'mean'), mean_sharpe=('sharpe', 'mean'))
           .reset_index())
summary
```
